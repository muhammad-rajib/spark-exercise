### PySpark DataFrames
* filter operation
* &,|,==
* ~

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession \
        .builder \
        .appName('filter_operation') \
        .getOrCreate()
```

```python
df = spark.read.csv('test_dataset_03.csv', header=True, inferSchema=True)
df.show()
```

```python
# filter: salary of the people less than 20000
df.filter(df['Salary'] <= 20000).show()

# way no. 2
df.filter("Salary <= 20000").show()
```

```python
# apply filter on selected columns
df.filter(df['Salary'] <= 20000).select(['Name', 'Age']).show()
```

```python
# range filtering: using '&'
df.filter((df['Salary'] <= 20000) & (df['Salary'] > 15000)).show()
```

```python
# not ~
df.filter(~(df['Salary'] == 20000)).show()
```

```python
# or |
df.show()
df.filter((df['Salary'] < 20000) | (df['Salary'] > 30000)).show()
```
