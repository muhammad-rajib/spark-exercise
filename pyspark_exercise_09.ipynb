### DataFrame
    * GroupBy
    * Aggregate Functions

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession \
        .builder \
        .appName('groupby_and_aggregateFunction') \
        .getOrCreate()
```

```python
df = spark.read.csv('test_dataset_04.csv', header=True, inferSchema=True)

df.show()
df.printSchema()
```

```python
# GroupBy
df.groupBy('Departments').sum().show()
```

```python
# mean()
df.groupBy('Departments').mean().show()
```

```python
# min()
df.show()
df.groupBy('Departments').min().show()
```
