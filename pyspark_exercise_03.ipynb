```python
import os
os.environ["PYARROW_IGNORE_TIMEZONE"] = "1"

import pyspark.pandas as ps
```

```python
psdf = ps.range(10)
type(psdf)
```

```python
pdf = psdf.to_pandas()
type(pdf)
```

```python
pdf.values
```

```python
psdf = ps.from_pandas(pdf)
type(psdf)
```

```python
psdf
```

```python
sdf = psdf.to_spark()
sdf.show()
```

```python
sdf.filter('id > 5').show()
```

```python
# convert Spark DataFrame to pandas-on-spark DataFrame
psdf = sdf.to_pandas_on_spark()
type(psdf)
```
